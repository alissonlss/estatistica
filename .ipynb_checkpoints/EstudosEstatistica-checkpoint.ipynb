```python
# Inportando bibliotecas base para visualização e analise de dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/aliss/Downloads/discentes-2019.csv", sep=";")
```

```python
df
```

### Vamos verificar a quantidade de dados nulos 

```python
df.info()
```

```python
df.isnull().sum()
```

Como os dados nulos estão nos dados que iremos utilizar, iremos retirá-los, pois eles podem interferir nas nossas analises.

Os campos que seram usados para as analises a seguir serão sexo, status, nível de ensino. Em analises futuras iremos relacionar os dados de status e curso e status e modalidade de ensino, para analisar se existe uma relação da situação que o dicente se encontra e suas condições, então para essas analises, será necessário realizar uma limpeza nos dados.

### Removendo dados nulos

```python
# A função dropna remove todos os dados nulos (nan), quando não especificado o eixo, ele remove as linhas.
df = df.dropna()
df
```

Vamos validar que os dados nulos foram removidos, para isso vamos dar um novo df.info()

```python
df.info()
```

```python
# Vamos reorganizar os index do nosso dataframe
def reorganizar_index():
    tam = len(df.index)
    df.index = np.arange(0,tam)
reorganizar_index()
df
```

### Iniciando as analises

Antes de iniciar as analises iremos obter quais são os dados que possuem no dataset, para isso, iremos pegar o nome das colunas para separar quais iremos analisar.

```python
columns = list(df.columns)
columns
```

#### - Vamos começar analisando o percentual de discentes por sexo:

```python
type_sexo = list(df['sexo'].unique())
sexo = df['sexo'].value_counts()

sexo
```

```python
# Iremos remover esse tipo false, pois não sabemos do que se trata
type_sexo = type_sexo[:-1]
sexo.pop('false')
sexo
```

```python
# Também iremos remover do dataframe e realinhar o index
df = df[df['sexo'] != 'false']
reorganizar_index()
```

Depois de ter retirado os sexos false, iremos mapear o percentual de discentes por sexo.

```python
fig, ax = plt.subplots()

# Criando o gráfico de setores
ax.pie(sexo, labels=sexo.index, autopct="%1.1f%%", colors=['orange', "pink"], textprops=dict(weight="600"))
ax.set_title("Distribuição de Discentes por Sexo:")

plt.show()
```

Como foi possível observar, o número de alunos ingressantes em 2019 na UFRN é a maioria do sexo masculino.

#### - Analise da quantidade de discente que se encontra em cada status:

```python
type_status = list(df['status'].unique())
count_status = df['status'].value_counts()
type_status
```

```python
count_status
```

```python
colors = ['red', 'green','blue', 'yellow', 'pink', 'cyan', 'orange', 'gray']

fig, ax = plt.subplots()

# Construindo o gráfico de barras verticais 
for i in range(len(type_status)):
    ax.bar(type_status[i][:3], count_status[i], color=colors[i])
ax.set_xlabel('Sintuação dos Discentes')
ax.set_ylabel('Quantidade dos Discentes')
ax.set_title('Quantidade dos Discentes por Situação Estudantil')
ax.legend(type_status)

plt.show()
```

Podemos obsevar que o número de discentes cancelados (5428) é maior do que o número de ativos (ativo + ativo_formando = 5019 + 238 = 5257). O que é um dado bem preocupande para o ambiente educacional.

### Analises em relação aos discentes cancelados:

Agora vamos separar os dados dos discentes cujo status está como cancelado e realizar analises em cima desses dados.

```python
# Vamos saparar apenas os discentes que estão na sintuação de cancelados
df_cancelado = df[df['status'] == 'CANCELADO']
df_cancelado
```

#### - Analise da distribuição de discentes cancelados por sexo:

Ápos a separação de discentes cancalados por sexo, iremos contabilizar e plotar um gráfico para identificar o percentual por sexo. 

```python
cancelado_sexo = df_cancelado['sexo'].value_counts()
cancelado_sexo
```

```python
fig, ax = plt.subplots()

# Criando o gráfico de setores
ax.pie(cancelado_sexo, labels=cancelado_sexo.index, autopct="%1.1f%%", colors=['orange', "pink"], textprops=dict(weight="600"))
ax.set_title("Distribuição de Discentes Cancelados por Sexo:")

plt.show()
```

Como na analise apenas por sexo, o número de discentes de sexo masculino teve a poncentagem maior em reação ao dados de cancelamento, o que representa quase ⅔ do número de discentes cancelados.

#### - Analise da distribuição de discentes cancelados por nível ensino:

```python
cancelado_ns = df_cancelado['nivel_ensino'].value_counts()
cancelado_ns
```

```python
fig, ax = plt.subplots(figsize=(6,6))

legend = ['GRADUAÇÃO', 'TÉCNICO', 'F. COMPLEMENTAR', 'LATO SENSU', 'MESTRADO', 'DOUTORADO', 'TÉC. INTEGRADO', 'RESIDÊNCIA']
explode = (0, 0, 0, 0, 0, 0, 0.5, 0.2)  

# Criando o gráfico de setores e reaproveitando as corres definida para o gráfico de barras
ax.pie(cancelado_ns, explode=explode, autopct="%1.1f%%", colors=colors, textprops=dict(weight="800"))
ax.set_title("Distribuição de Discentes Cancelados por Sexo:")
ax.legend(legend, loc="upper right")

plt.show()
```

Podemos obsevar que tivemos um número maior de cancelamento de alunos de graduação.

#### - Analise da distribuição de discentes cancelados por modalidade educacional:

```python
cancelado_mod_edu = df_cancelado['modalidade_educacao'].value_counts()
cancelado_mod_edu
```

```python
fig, ax = plt.subplots(figsize=(6,6)) 

colors = ["c", "r", "y"]

# Criando o gráfico de setores
ax.pie(cancelado_mod_edu, labels=cancelado_mod_edu.index, autopct="%1.1f%%", colors=colors, textprops=dict(weight="600"))
ax.set_title("Distribuição de Discentes Cancelados pela Modalidade de Ensino:")

plt.show()
```

Como podemos obsevar no gráfico anterior, o maior número de discentes cancelados cursavam a modalidade de ensino de forma presencial.

### Analise dos cursos de TI:

```python
# Palavras chave para separar os cursos
ti = ['SISTEMA', 'COMPUTAÇÃO', 'SOFTWARE', 'TECNOLOGIA', 'DIGITAL']

def curso_ti(df, coluna):
    cont_alunos = 0
    cursos_ti = []

    for i in df[coluna]:
        for j in ti:
            if (j in i):
                cont_alunos += 1
                if (i not in cursos_ti):
                    cursos_ti.append(i)
                break
    return cont_alunos, cursos_ti
```

#### - Analise da quantidade de discentes dos cursos de TI:

```python
dados_ti = curso_ti(df, 'nome_curso')

quant_alunos_ti = dados_ti[0]
cursos_ti = dados_ti[1]

print(f'Quantidade de alunos em TI: {quant_alunos_ti}')
```

Lista de cursos de TI:

```python
cursos_ti
```

```python
num_total = df.shape[0]
perc_ti = np.around((quant_alunos_ti*100)/num_total, 2)

print(f"OS discentes ingressantes em TI representam {perc_ti}%.")
```

#### - Analise dos cursos de TI que tiveram dicentes cancelados:

```python
cancelados_ti = curso_ti(df_cancelado, 'nome_curso')

quant_alunos_c = cancelados_ti[0]
cursos_c = cancelados_ti[1]

print(f'Quantidade de alunos cancelados em TI: {quant_alunos_c}')
```

Lista de cursos de TI com alunos cancelados:

```python
cursos_c
```

```python
num_total = df_cancelado.shape[0]
perc_ti_canc = np.around((quant_alunos_c*100)/num_total, 2)

print(f"Discente cancelados em TI representam {perc_ti_canc}%.")
```
